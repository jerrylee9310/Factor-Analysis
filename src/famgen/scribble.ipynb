```python
from famgam.core import FamGenoSimul, GenotypeValidator
```

# Generate Family Genotype

```python
fam_geno = FamGenoSimul(num_variant=1000)
fam_geno.generate_parent_haplotype(num_sample=100)
fam_geno.make_offspring_haplotype(num_offspring=2)
```

# Get genotype as `DataFame`

```python
# 모든 구성원의 genotype을 DataFrame으로 가져오기
raw_geno = fam_geno.get_genotype()
raw_geno
# # 특정 역할의 genotype만 가져오기
# father_genotypes_df = simulator.get_genotype(role="father")

# # numpy array 형태로 가져오기
# all_genotypes_array = simulator.get_genotype(as_dataframe=False)
```

```python
# Standardized genotype 얻기
std_geno = fam_geno.get_genotype(std=True)
std_geno
# # 특정 역할의 standardized genotype 얻기
# father_std_geno = simulator.get_genotype(role="father", std=True)
```

# Validataion

```python
# MAF
validator = GenotypeValidator()
maf_results = validator.validate_maf(
    genotypes=(fam_geno
               .get_genotype(as_dataframe=True, std=False)
               .drop(["family_id", "role"], axis=1)
               .values),
    maf=fam_geno.haplotype_generator.maf,
    plot=False
)
```

```python
# correlation
corr_results = validator.validate_correlation(
    fam_geno.get_genotype(as_dataframe=True, std=True),
    plot=False
)
```
